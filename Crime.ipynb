## Part 5 : Dimension Reduction

```python
X = NEW_DF_Crime.iloc[:, 2:]
y = NEW_DF_Crime.iloc[:, 0]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y ,train_size=0.70, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA

c = 7
pca = PCA(n_components = c)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

plt.scatter(np.arange(1,(c+1)),pca.explained_variance_, c = 'red')
plt.plot((0,c), (1,1), color = 'black', linestyle = 'dashed')
plt.xlabel('PC')
plt.ylabel('Amount of variance explained')
plt.show()
print(X_train.shape)
```

## Part 6 : Regression Models (Testing with Cross Validation)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import make_scorer

kfold = KFold(n_splits=10, random_state=123, shuffle = True)

results=[]
names=[]
R2=[]
MSE=[]
MAE=[]
```

### 6-1: Linear Regression

```python
from sklearn.linear_model import LinearRegression
LM = LinearRegression()
LM_CV = cross_val_score(LM, X_train, y_train, cv=kfold, scoring= make_scorer(r2_score))
results.append(LM_CV)
names.append('LM')
LM_CV.mean()
```

```python
LM.fit(X_train, y_train)
y_pred = LM.predict(X_train)
R2.append(r2_score(y_train,y_pred))
MSE.append(mean_squared_error(y_train,y_pred))
MAE.append(mean_absolute_error(y_train,y_pred))
print('Linear Regression:')
print('R^2: %.4f' % r2_score(y_train,y_pred))
print('MSE: %.4f' % mean_squared_error(y_train,y_pred))
print('MAE: %.4f' % mean_absolute_error(y_train,y_pred))
```

### 6-2: Linear Regression

```python
from sklearn.svm import SVR
SVRM = SVR()
SVR_CV = cross_val_score(SVRM, X_train, y_train, cv=kfold, scoring = make_scorer(r2_score))
results.append(SVR_CV)
names.append('SVR')
SVR_CV.mean()
```

```python
SVRM.fit(X_train, y_train)
y_pred = SVRM.predict(X_train)
R2.append(r2_score(y_train,y_pred))
MSE.append(mean_squared_error(y_train,y_pred))
MAE.append(mean_absolute_error(y_train,y_pred))
print('Support Vector Regression:')
print('R^2: %.4f' % r2_score(y_train,y_pred))
print('MSE: %.4f' % mean_squared_error(y_train,y_pred))
print('MAE: %.4f' % mean_absolute_error(y_train,y_pred))
```

### 6-3: Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF_CV = cross_val_score(RF, X_train, y_train, cv=kfold, scoring = make_scorer(r2_score))
results.append(RF_CV)
names.append('RF')
RF_CV.mean()
```

```python
RF.fit(X_train, y_train)
y_pred = RF.predict(X_train)
R2.append(r2_score(y_train,y_pred))
MSE.append(mean_squared_error(y_train,y_pred))
MAE.append(mean_absolute_error(y_train,y_pred))
print('Random Forest Regression:')
print('R^2: %.4f' % r2_score(y_train,y_pred))
print('MSE: %.4f' % mean_squared_error(y_train,y_pred))
print('MAE: %.4f' % mean_absolute_error(y_train,y_pred))
```

### 6-4: Ridge Regression

```python
from sklearn.linear_model import Ridge
RIDGE = Ridge(alpha=1.0)
RIDGE_CV = cross_val_score(RIDGE, X_train, y_train, cv=kfold, scoring = make_scorer(r2_score))
results.append(RIDGE_CV)
names.append('Ridge')
RIDGE_CV.mean()
```

```python
RIDGE.fit(X_train, y_train)
y_pred = RIDGE.predict(X_train)
R2.append(r2_score(y_train,y_pred))
MSE.append(mean_squared_error(y_train,y_pred))
MAE.append(mean_absolute_error(y_train,y_pred))
print('Ridge Regression:')
print('R^2: %.4f' % r2_score(y_train,y_pred))
print('MSE: %.4f' % mean_squared_error(y_train,y_pred))
print('MAE: %.4f' % mean_absolute_error(y_train,y_pred))
```

### 6-5: Lasso Regression

```python
from sklearn.linear_model import Lasso
LassoR = Lasso(alpha=0.01)
LassoR_CV = cross_val_score(LassoR, X_train, y_train, cv=kfold, scoring = make_scorer(r2_score))
results.append(LassoR_CV)
names.append('Lasso')
LassoR_CV.mean()
```

```python
LassoR.fit(X_train, y_train)
y_pred = LassoR.predict(X_train)
R2.append(r2_score(y_train,y_pred))
MSE.append(mean_squared_error(y_train,y_pred))
MAE.append(mean_absolute_error(y_train,y_pred))
print('Lasso Regression:')
print('R^2: %.4f' % r2_score(y_train,y_pred))
print('MSE: %.4f' % mean_squared_error(y_train,y_pred))
print('MAE: %.4f' % mean_absolute_error(y_train,y_pred))
```

## Part 7 : Regressor Comparison

```python
fig = plt.figure()
fig.suptitle('Regressor Comparison - Cross Validation')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
%matplotlib inline
fig = plt.figure()
fig.suptitle('Regressor Comparison - (R^2, MSE, MAE)')
ax = fig.add_subplot(111)
plt.plot(R2, 'ro',MSE,'b+',MAE,'gv')
plt.xlim(0,5)
plt.ylim(-.1,1)

ax.set_xticklabels(names)

from matplotlib.lines import Line2D
colors = ['red', 'blue', 'green']
lines = [Line2D([0], [0], color=c, linewidth=3, linestyle='--') for c in colors]
labels = ['R2', 'MSE', 'MAE']
plt.legend(lines, labels)

plt.show()
```
